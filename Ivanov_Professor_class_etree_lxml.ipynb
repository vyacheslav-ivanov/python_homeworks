```python
import re
import requests
from lxml import html
```

```python
class Professor:
    def __init__(self, fullname, workplace, interest, telnum):
        self.fullname = fullname
        self.workplace = workplace
        self.interest = interest
        self.telnum = telnum

```

```python
def teachers_with_xpath(tree):
    fullname = tree.xpath('//a[@class="link link_dark large b"]/text()')
    workplace = tree.xpath('//a[@class="link"]/text()')
    interest = tree.xpath('//a[@class="tag"]/text()')
    telnum = tree.xpath('//div[@class="l-extra small"]//span[1]/text()')
    for i in fullname:
        if i == '\n\t\t\t\t\t':
            fullname.remove(i)

            
    staff = []
    for n, w, i, t in zip(fullname, workplace, interest, telnum):
        professor = Professor(n.strip('\n\t\t\t\t\t'), w, i, t)
        staff.append(professor)
    return staff
    
```

```python
page = requests.get('https://www.hse.ru/org/persons/?ltr=%D0%94;udept=22726')
text = page.content.decode('utf-8')
tree = html.fromstring(text)
for j in teachers_with_xpath(tree):
    print('ФИО:', j.fullname)
    print('Место работы:', j.workplace)
    print('Интересы:', j.interest)
    print('Номер телефона:', j.telnum)
    print()
```
